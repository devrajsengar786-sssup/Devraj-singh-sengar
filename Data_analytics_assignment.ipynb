**#Case Study 1: Student Marks Analysis**

 Problem Statement:

Analyze student performance using NumPy.


```python
# Step 1: Import Library
import numpy as np
```

```python
# Step 2: Dataset
marks = np.array([
    [80, 75, 90],
    [60, 88, 70],
    [95, 92, 85],
    [70, 65, 60],
    [88, 78, 84]
])

print("Marks Dataset:\n", marks)
```

```python
# Task 1: Total Marks of Each Student
total_marks = np.sum(marks, axis=1)
print("Total Marks:", total_marks)
```

```python
# Task 2: Average Marks Per Student
average_marks = np.mean(marks, axis=1)
print("Average Marks:", average_marks)
```

```python
# Task 3: Identify Topper
topper_index = np.argmax(total_marks)
print("Topper is Student", topper_index + 1)
```

```python
# Task 4: Subject-wise Highest Score
subject_highest = np.max(marks, axis=0)
print("Highest Score in Each Subject:", subject_highest)
```

```python
# Task 5: Standard Deviation of Each Subject
subject_std = np.std(marks, axis=0)
print("Standard Deviation of Each Subject:", subject_std)
```

**#Interpretation (Student Performance)**

1.Student 3 has the highest total marks, so they are the topper.

2.Subject-wise highest marks show Subject 1 and 2 have very high performers.

3.Standard deviation indicates moderate variation among students.

4.Students with consistent averages show balanced performance across subjects.


**#Case Study 3: Sales Analysis**

Problem Statement:

Analyze weekly sales performance.

```python
# Dataset
sales = np.array([1200, 1500, 1100, 1700, 1600, 1800, 1400])
print("Sales Data:", sales)
```

```python
# Task 1: Average Daily Sales
average_sales = np.mean(sales)
print("Average Daily Sales:", average_sales)
```

```python
# Task 2: Day with Highest Sales
highest_day = np.argmax(sales)
print("Highest Sales on Day", highest_day + 1)
```

```python
# Task 3: Percentage Growth (Day 1 to Day 7)
growth = ((sales[-1] - sales[0]) / sales[0]) * 100
print("Percentage Growth:", growth, "%")
```

```python
# Task 4: Cumulative Sales
cumulative_sales = np.cumsum(sales)
print("Cumulative Sales:", cumulative_sales)
```

```python
# Task 5: Standard Deviation
sales_std = np.std(sales)
print("Standard Deviation:", sales_std)
```

**#Interpretation (Sales Pattern)**

1.The average daily sales give an overall business performance measure.

2.The highest sales occurred on Day 6, showing peak customer activity.

3.There is a positive growth from Day 1 to Day 7, indicating improvement.

4.Standard deviation shows moderate fluctuation in daily sales.

5.Cumulative sales show steady revenue growth across the week.
