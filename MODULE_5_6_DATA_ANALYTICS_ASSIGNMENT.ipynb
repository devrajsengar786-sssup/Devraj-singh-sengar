## MODULE 6 – STATISTICAL ANALYSIS

```python
# 1. T-Test
from scipy.stats import ttest_ind

low_exp = df[df['YearsExperience'] < 5]['Salary']
high_exp = df[df['YearsExperience'] >= 5]['Salary']

t_stat, p_val = ttest_ind(low_exp, high_exp)

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

**INTERPRETATION**

p < 0.05 → significant difference in salaries

Usually: High experience group earns significantly more

```python
# 2. Chi-Square Test
from scipy.stats import chi2_contingency

df['Salary_Category'] = pd.cut(df['Salary'], bins=3, labels=['Low','Medium','High'])

table = pd.crosstab(df['Salary_Category'], df['YearsExperience'] > 5)

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square:", chi2)
print("P-value:", p)
```

**INTERPRETATION**

Checks relationship between experience level and salary category

**3. P-Value Explanation**

p-value shows significance of result

p < 0.05 → reject null hypothesis

p > 0.05 → accept null hypothesis
