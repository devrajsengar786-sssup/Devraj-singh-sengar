## Task 1: Array Creation(1 Mark)

Create:

1D array of 10 integers

2D array (3×3 matrix)

Identity matrix of size 4

Random array (5×5)

## Task 2: Indexing & Slicing (1 Mark)

Given:

A = np.arange(1, 26).reshape(5,5)

Perform:

Extract 2nd row

Extract 3rd column

Extract 2×2 submatrix (center)

Reverse last row

```python
import numpy as np

A = np.arange(1, 26).reshape(5,5)
print("Matrix A:")
print(A)

second_row = A[1]
print("\n2nd Row:")
print(second_row)

third_column = A[:, 2]
print("\n3rd Column:")
print(third_column)

center_matrix = A[1:3, 1:3]
print("\n2x2 Center Matrix:")
print(center_matrix)

reversed_last_row = A[-1][::-1]
print("\nReversed Last Row:")
print(reversed_last_row)
```

## Task 3: Broadcasting(1 Mark)

Perform broadcasting operations on matrix A created in Task-2.

Add scalar 10 to matrix A

Add vector [1,2,3,4,5] to each row

Normalize matrix values between 0–1

```python
import numpy as np

A = np.arange(1, 26).reshape(5,5)
print("Original Matrix A:")
print(A)

add_scalar = A + 10
print("\nAfter Adding Scalar 10:")
print(add_scalar)

vector = np.array([1, 2, 3, 4, 5])
add_vector = A + vector
print("\nAfter Adding Vector [1,2,3,4,5] to each row:")
print(add_vector)

normalized = (A - A.min()) / (A.max() - A.min())
print("\nNormalized Matrix (0–1 range):")
print(normalized)
```

## Task 4: Vectorized Computation(1 Marks)

Perform element-wise multiplication
Compute dot product
Compute mean, standard deviation

```python
import numpy as np

X = np.array([1,2,3,4])
Y = np.array([5,6,7,8])

elementwise = X * Y
print("Element-wise Multiplication:")
print(elementwise)

dot_product = np.dot(X, Y)
print("\nDot Product:")
print(dot_product)

mean_X = np.mean(X)
mean_Y = np.mean(Y)
print("\nMean of X:", mean_X)
print("Mean of Y:", mean_Y)

std_X = np.std(X)
std_Y = np.std(Y)
print("\nStandard Deviation of X:", std_X)
print("Standard Deviation of Y:", std_Y)
```

Task 5: Mini Case Problem (1 Mark)

Dataset: Student marks (5 students, 3 subjects)

Compute total and average per student

Find topper

Find subject-wise highest score

```python
import numpy as np

marks = np.array([
    [80,75,90],
    [60,88,70],
    [95,92,85],
    [70,65,60],
    [88,78,84]
])

print("Marks Matrix:")
print(marks)

total = np.sum(marks, axis=1)
print("\nTotal Marks per Student:")
print(total)

average = np.mean(marks, axis=1)
print("\nAverage Marks per Student:")
print(average)

topper_index = np.argmax(total)
print("\nTopper Student Index (0-based):", topper_index)
print("Topper Total Marks:", total[topper_index])

subject_highest = np.max(marks, axis=0)
print("\nSubject-wise Highest Scores:")
print(subject_highest)
```
