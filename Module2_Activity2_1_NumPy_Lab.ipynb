## Task 2: Indexing & Slicing (1 Mark)

Given:

A = np.arange(1, 26).reshape(5,5)

Perform:

Extract 2nd row

Extract 3rd column

Extract 2×2 submatrix (center)

Reverse last row

```python
A = np.arange(1, 26).reshape(5,5)
print("MATRIX A")
print(A)

# Extract 2nd row

row2 = A[1]
print("\n2nd Row:")
print(row2)

# Extract 3rd column

column3 = A[:, 2]
print("\n3rd Column:")
print(column3)

# Extract 2x2 center matrix

center_matrix = A[1:3, 1:3]
print("\n2x2 Center Matrix:")
print(center_matrix)

# Reverse last row
reversed_last_row = A[-1][::-1]
print("\nReversed Last Row:")
print(reversed_last_row)
```

## Task 3: Broadcasting(1 Mark)

Perform broadcasting operations on matrix A created in Task-2.

Add scalar 10 to matrix A

Add vector [1,2,3,4,5] to each row

Normalize matrix values between 0–1

```python
# Perform broadcasting operations here

a = np.arange(1, 26).reshape(5,5)
print("Original Matrix A:")
print(a)

# Add scalar 10 to matrix

add = a + 10
print("\nAfter Adding Scalar 10:")
print(add)

# Add vector [1,2,3,4,5] to each row

vector = np.array([1, 2, 3, 4, 5])
add_vector = a + vector
print("\nAfter Adding Vector [1,2,3,4,5] to each row:")
print(add_vector)

# Normalize matrix between 0 and 1
normalized = (a - a.min()) / (a.max() - a.min())
print("\nNormalized Matrix (0–1 range):")
print(normalized)
```

## Task 4: Vectorized Computation(1 Marks)

Perform element-wise multiplication
Compute dot product
Compute mean, standard deviation

```python
X = np.array([1,2,3,4])
Y = np.array([5,6,7,8])

# Element-wise multiplication

multiply = X * Y
print("Element-wise Multiplication:")
print(multiply)

# Dot product
product = np.dot(X, Y)
print("\nDot Product:")
print(product)
# Mean and Standard Deviation

mean_X = np.mean(X)
mean_Y = np.mean(Y)
print("\nMean of X:", mean_X)
print("Mean of Y:", mean_Y)

standard_X= np.std(X)
standard_Y = np.std(Y)
print("\nStandard Deviation of X:", standard_X)
print("Standard Deviation of Y:",standard_Y )
```

Task 5: Mini Case Problem (1 Mark)

Dataset: Student marks (5 students, 3 subjects)

Compute total and average per student

Find topper

Find subject-wise highest score

```python
marks = np.array([[80,75,90],[60,88,70],[95,92,85],[70,65,60],[88,78,84]])
# Compute totals, averages, topper
print("marks matrix")
print(marks)

total= np.sum(marks,axis=1)
print("\ntotal marks")
print(total)

average=np.mean(marks,axis=1)
print("\naverage marks")
print(average)

topper=np.argmax(total)
print("\ntopper student marks",total[topper])
```
