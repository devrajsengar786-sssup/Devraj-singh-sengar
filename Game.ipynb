```python
import random

def choose_level():
    print("\nChoose Difficulty Level:")
    print("1. Easy (10 lives)")
    print("2. Medium (7 lives)")
    print("3. Hard (5 lives)")

    while True:
        choice = input("Enter 1 / 2 / 3: ")
        if choice == "1":
            return 10
        elif choice == "2":
            return 7
        elif choice == "3":
            return 5
        else:
            print("Invalid choice. Try again.")

def play_game():
    number = random.randint(1, 100)
    lives = choose_level()
    score = 0
    previous_diff = None

    print("\nğŸ² I have chosen a number between 1 and 100.")
    print("Try to guess it!\n")

    while lives > 0:
        try:
            guess = int(input("Enter your guess: "))
        except ValueError:
            print("âŒ Please enter a valid number.")
            continue

        if guess < 1 or guess > 100:
            print("âš ï¸ Guess must be between 1 and 100.")
            continue

        diff = abs(number - guess)

        if guess == number:
            score = lives * 10
            print("\nğŸ‰ BOOM! You guessed it right!")
            print(f"ğŸ† Your Score: {score}")
            if lives == 5:
                print("ğŸ˜ Perfect strategy! Are you a mind reader?")
            break

        lives -= 1

        # Hot & Cold hints
        if diff <= 5:
            print("ğŸ”¥ SUPER HOT!")
        elif diff <= 10:
            print("â™¨ï¸ Warm!")
        elif diff <= 20:
            print("â„ï¸ Cold!")
        else:
            print("ğŸ§Š Freezing!")

        # Smarter hint
        if previous_diff is not None:
            if diff < previous_diff:
                print("ğŸ“ˆ You're getting closer!")
            else:
                print("ğŸ“‰ You're moving away!")

        previous_diff = diff

        # Direction hint
        if guess < number:
            print("â¬†ï¸ Try a higher number.")
        else:
            print("â¬‡ï¸ Try a lower number.")

        print(f"â¤ï¸ Lives left: {lives}\n")

    else:
        print("\nğŸ’€ Game Over!")
        print(f"The correct number was: {number}")

def main():
    print("ğŸ® WELCOME TO THE ULTIMATE NUMBER GUESSING GAME ğŸ®")

    while True:
        play_game()
        again = input("\nDo you want to play again? (y/n): ").lower()
        if again != "y":
            print("\nğŸ‘‹ Thanks for playing! See you next time.")
            break

main()

```
